```
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.20, random_state=42, stratify=y_final)
numerical_features = X_final.select_dtypes(include=np.number).columns.tolist()
categorical_features = X_final.select_dtypes(exclude=np.number).columns.tolist()
preprocessor_linear = ColumnTransformer(transformers=[('num', StandardScaler(), numerical_features), ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)])
preprocessor_tree = ColumnTransformer(transformers=[('num', StandardScaler(), numerical_features), ('cat', TargetEncoder(), categorical_features)])
preprocessor_nb = ColumnTransformer(
    transformers=[
        ('num', KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='uniform'), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)
model_results = {}
print("✅ Data splitting and preprocessing pipelines are ready.")
```

## 7. Broad Model Search with Optuna Baseline Tuning
This is the first, crucial stage of our two-stage tuning process. We perform a Bayesian optimization search for **every single models** we are using to ensure we are comparing the strongest possible versions of each algorithm to establish a robust and reliable performance baseline.

### 7.1 Optuna Baseline Objective Function

```
BASELINE_OPTUNA_TRIALS = 25 # Number of trials for the initial search for each model

def run_baseline_study(model_name, model, params_lambda, preprocessor):
    """A generic function to run a baseline Optuna study for any given model."""
    print(f"--- Tuning {model_name} ---")
    start_time = time.time()

    # ----------------------------------------------------------------------
    # FIX: Dynamic determination of n_jobs based on likely GPU usage or known memory issues
    # ----------------------------------------------------------------------
    cv_n_jobs = -1 # Default to using all cores

    # 1. Check if the model is configured for GPU execution (requires n_jobs=1 for stability)
    #    We check the instantiation parameters, assuming the global 'is_gpu_available'
    #    flag was used to set 'task_type' (CatBoost), 'tree_method' (XGBoost), or 'device' (LGBM).

    # Check CatBoost
    if model_name == 'CatBoost' and model.get_params().get('task_type') == 'GPU':
        cv_n_jobs = 1
        print("  [INFO] Setting CV n_jobs=1: CatBoost is configured for GPU (TerminatedWorkerError mitigation).")

    # Check XGBoost
    elif model_name == 'XGBoost' and model.get_params().get('tree_method') == 'gpu_hist':
        cv_n_jobs = 1
        print("  [INFO] Setting CV n_jobs=1: XGBoost is configured for GPU (TerminatedWorkerError mitigation).")

    # Check LightGBM
    elif model_name == 'LightGBM' and model.get_params().get('device') == 'gpu':
        cv_n_jobs = 1
        print("  [INFO] Setting CV n_jobs=1: LightGBM is configured for GPU (TerminatedWorkerError mitigation).")

    # 2. Add a general safeguard for any GPU availability when using memory-intensive steps
    #    This catches any non-CatBoost/XGBoost/LGBM model that might use the GPU
    #    (though less common) or if the OOM is purely from SMOTEENN on a high-core count CPU.
    elif is_gpu_available:
        # Safety check: If a GPU is present, and we're using complex pipelines, sequence CV.
        cv_n_jobs = 1
        print("  [INFO] Setting CV n_jobs=1: GPU is available, running CV sequentially for stability.")

    else:
        # Default: Use all cores (n_jobs=-1) for CPU-only execution
        cv_n_jobs = -1
        print("  [INFO] Setting CV n_jobs=-1: Using all CPU cores for cross-validation.")

    # ----------------------------------------------------------------------

    def objective(trial):
        params = params_lambda(trial)
        clf = model.set_params(**params)
        pipeline = ImbPipeline([
            ('preprocessor', preprocessor),
            ('sampler', SMOTEENN(random_state=42)),
            ('classifier', clf)
        ])

        # Use the conditionally set n_jobs variable
        return cross_val_score(pipeline, X_train, y_train, n_jobs=cv_n_jobs, cv=3, scoring='roc_auc').mean()

    study = optuna.create_study(direction='maximize')
    study.optimize(objective, n_trials=BASELINE_OPTUNA_TRIALS)

    end_time = time.time()
    print(f"Best CV ROC-AUC: {study.best_value:.4f} (Completed in {end_time - start_time:.2f}s)")
    print(f"Best Params here are: {study.best_params}")
    model_results[model_name] = (study.best_value, study.best_params)
```

## 8. Advanced Deep Learning Architectures
The below code prepares training and testing datasets for a machine learning model by:

1. **Encoding categorical features** as integers (`cat.codes`).
2. **Scaling numerical features** to have a mean of 0 and standard deviation of 1 using `StandardScaler`.
3. **Storing categorical feature dimensions** (the number of unique categories) in a dictionary (`cat_dims`).
4. Creating dictionaries (`train_inputs` and `test_inputs`) that combine both categorical and numerical features, ready for model input.

It essentially preprocesses the data by handling both categorical and numerical features in a way suitable for a deep learning(Keras) model.


```
X_train_keras = X_train.copy()
X_test_keras = X_test.copy()
cat_dims = {}
for col in categorical_features:
    X_train_keras[col] = X_train_keras[col].astype('category').cat.codes
    X_test_keras[col] = X_test_keras[col].astype('category').cat.codes
    cat_dims[col] = len(X_final[col].astype('category').cat.categories)
scaler = StandardScaler()
X_train_keras[numerical_features] = scaler.fit_transform(X_train_keras[numerical_features])
X_test_keras[numerical_features] = scaler.transform(X_test_keras[numerical_features])
train_inputs = {col: np.array(X_train_keras[col]) for col in categorical_features} | {'numerical_input': X_train_keras[numerical_features].values}
test_inputs = {col: np.array(X_test_keras[col]) for col in categorical_features} | {'numerical_input': X_test_keras[numerical_features].values}
```

### 8.1 Baseline Keras NN
A basic feedforward neural network with one or more dense (fully connected) layers, often starting with a single hidden layer using ReLU activation and an output layer suited to the task (e.g., sigmoid for binary classification).

```
def build_baseline_keras_model():
    inputs, embeddings = [], []
    for col in categorical_features:
        input_cat = Input(shape=(1,), name=col); inputs.append(input_cat)
        embedding_size = min(np.ceil(cat_dims[col]/2), 100)
        embedding = Flatten()(Embedding(cat_dims[col], int(embedding_size))(input_cat)); embeddings.append(embedding)
    input_num = Input(shape=(len(numerical_features),), name="numerical_input"); inputs.append(input_num)
    x = Concatenate()(embeddings + [input_num])
    x = Dense(128, activation='relu')(x); x = Dropout(0.3)(x)
    x = Dense(64, activation='relu')(x); x = Dropout(0.3)(x)
    output = Dense(1, activation='sigmoid')(x)
    return Model(inputs=inputs, outputs=output)

keras_model = build_baseline_keras_model()
keras_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
keras_model.fit(train_inputs, y_train, epochs=1000, batch_size=64, validation_split=0.2, callbacks=[EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)], verbose=0)
y_proba_keras = keras_model.predict(test_inputs).ravel()
keras_roc_auc = roc_auc_score(y_test, y_proba_keras)
print(f"--- Baseline Keras NN Test ROC-AUC: {keras_roc_auc:.4f} ---")
model_results['Keras NN'] = (keras_roc_auc, {'architecture': 'Embedding-128-64-1'})
```

### 8.2 Wide & Deep Network
A model that combines a "wide" part (linear model that memorizes patterns using raw or one-hot encoded features) and a "deep" part (neural network that generalizes via embeddings and hidden layers), allowing it to learn both memorization and generalization simultaneously.

```
def build_wide_and_deep_model():
    inputs, embeddings = [], []
    for col in categorical_features:
        input_cat = Input(shape=(1,), name=col)
        embedding_size = min(np.ceil(cat_dims[col]/2), 100)
        embedding = Flatten()(Embedding(cat_dims[col], int(embedding_size))(input_cat)); embeddings.append(embedding)
        inputs.append(input_cat)
    input_num = Input(shape=(len(numerical_features),), name="numerical_input"); inputs.append(input_num)
    deep_path = Concatenate()(embeddings + [input_num])
    deep_path = Dense(128, activation='relu')(deep_path); deep_path = Dropout(0.3)(deep_path)
    deep_path = Dense(64, activation='relu')(deep_path)
    wide_path = Concatenate()(embeddings + [input_num])
    combined = Concatenate()([wide_path, deep_path])
    output = Dense(1, activation='sigmoid')(combined)
    return Model(inputs=inputs, outputs=output)

wide_deep_model = build_wide_and_deep_model()
wide_deep_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])
wide_deep_model.fit(train_inputs, y_train, epochs=1000, batch_size=64, validation_split=0.2, callbacks=[EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)], verbose=0)
y_proba_wd = wide_deep_model.predict(test_inputs).ravel()
wd_roc_auc = roc_auc_score(y_test, y_proba_wd)
print(f"--- Wide & Deep Model Test ROC-AUC: {wd_roc_auc:.4f} ---")
model_results['Wide & Deep'] = (wd_roc_auc, {'architecture': 'Wide&Deep-128-64-1'})
```

### 8.3 Autoencoder for Feature Extraction

The below code trains an **autoencoder**, a type of neural network that learns to compress and then reconstruct data, to reduce the original features into a smaller set of 32 meaningful ones. After training, it keeps only the compressed (encoded) features and feeds them into an **XGBoost classifier** to make predictions. The goal is to improve performance by using features automatically learned by the autoencoder, and the model's accuracy is evaluated using the ROC-AUC score on the test set.


```
input_dim = X_train_keras.shape[1]
encoding_dim = 32
input_layer = Input(shape=(input_dim,))
encoder = Dense(128, activation='relu')(input_layer)
encoder = Dense(64, activation='relu')(encoder)
encoder = Dense(encoding_dim, activation='relu')(encoder)
decoder = Dense(64, activation='relu')(encoder)
decoder = Dense(128, activation='relu')(decoder)
decoder = Dense(input_dim, activation='sigmoid')(decoder)
autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
autoencoder.fit(X_train_keras, X_train_keras, epochs=50, batch_size=64, shuffle=True, validation_split=0.2, verbose=0, callbacks=[EarlyStopping(monitor='val_loss', patience=10)])
encoder_model = Model(inputs=input_layer, outputs=encoder)
X_train_encoded = encoder_model.predict(X_train_keras)
X_test_encoded = encoder_model.predict(X_test_keras)
xgb_on_ae = XGBClassifier(n_estimators=1000, learning_rate=0.05, use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_on_ae.fit(X_train_encoded, y_train)
y_proba_ae = xgb_on_ae.predict_proba(X_test_encoded)[:, 1]
ae_roc_auc = roc_auc_score(y_test, y_proba_ae)
print(f"--- XGBoost on Autoencoder Features Test ROC-AUC: {ae_roc_auc:.4f} ---")
model_results['XGB on AE Features'] = (ae_roc_auc, {'architecture': 'AE(32)-XGB'})
```

## 10. Intensive Hyperparameter Tuning for Top Models
Here as there were 4 models with accuracy above 80 Percent, we are using the stacking classifier i.e. ensemble of these 4 models.

```
from sklearn.base import BaseEstimator, TransformerMixin
import pandas as pd


# --- 1. Configuration for Classical Models (tuning_config) ---
# =============================================================
tuning_config = {
    'Gradient Boosting': {'model': GradientBoostingClassifier(random_state=42), 'params': lambda trial: {'n_estimators': trial.suggest_int('n_estimators', 458, 460), 'learning_rate': trial.suggest_float('learning_rate', 0.245, 0.255, log=True), 'max_depth': trial.suggest_int('max_depth', 4, 8)} },
    #Updated the models to search for the best combinations from near the best parameters found earlier.
    'XGBoost': {
        'model': XGBClassifier(
            random_state=42,
            use_label_encoder=False,
            eval_metric='logloss',
            tree_method='gpu_hist' if is_gpu_available else 'auto'
        ),
        'params': lambda trial: {'n_estimators': trial.suggest_int('n_estimators', 416, 420), 'max_depth': trial.suggest_int('max_depth', 2, 4), 'learning_rate': trial.suggest_float('learning_rate', 0.0705, 0.0715, log=True), 'subsample': trial.suggest_float('subsample', 0.5, 1.0), 'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0)}
    },
    'LightGBM': {
        'model': LGBMClassifier(
            random_state=42,
            class_weight='balanced',
            device='gpu' if is_gpu_available else 'cpu'
        ),
        'params': lambda trial: {'n_estimators': trial.suggest_int('n_estimators', 470, 472), 'learning_rate': trial.suggest_float('learning_rate', 0.101, 0.104, log=True), 'num_leaves': trial.suggest_int('num_leaves', 72, 76), 'max_depth': trial.suggest_int('max_depth', 5, 25)}
    },
    'CatBoost': {
        'model': CatBoostClassifier(
            random_state=42,
            verbose=0,
            auto_class_weights='Balanced',
            task_type='GPU' if is_gpu_available else 'CPU'
        ),
        'params': lambda trial: {'iterations': trial.suggest_int('iterations', 268, 272), 'learning_rate': trial.suggest_float('learning_rate', 0.058, 0.062, log=True), 'depth': trial.suggest_int('depth', 3, 5)}
    }
}


def universal_sklearn_objective(trial, model_name):
    """
    Optuna objective function for all scikit-learn compatible models.
    """
    # Assumes 'tuning_config', 'ImbPipeline', 'preprocessor_tree', 'SMOTEENN',
    # 'X_train', and 'y_train' are defined in your notebook.
    config = tuning_config[model_name]
    model = config['model']
    params = config['params'](trial)
    model.set_params(**params)

    # Create a pipeline with preprocessing, sampling, and the classifier
    pipeline = ImbPipeline([
        ('preprocessor', preprocessor_tree),
        ('sampler', SMOTEENN(random_state=42)),
        ('classifier', model)
    ])

    # Evaluate the pipeline using cross-validation
    return cross_val_score(pipeline, X_train, y_train, cv=3, scoring='roc_auc', n_jobs=1).mean()

def create_baseline_model(n_units_1=128, n_units_2=64, dropout=0.3, optimizer='Adam', learning_rate=0.001):
    """
    Factory function to build the baseline Keras Neural Network with embeddings.
    """
    # Assumes 'categorical_features', 'cat_dims', 'numerical_features', 'Input',
    # 'Flatten', 'Embedding', 'Concatenate', 'Dense', 'Dropout', 'Model', and 'keras' are defined/imported.
    inputs, embeddings = [], []
    for col in categorical_features:
        input_cat = Input(shape=(1,), name=col)
        inputs.append(input_cat)
        embedding_size = min(np.ceil(cat_dims[col]/2), 50)
        embedding = Flatten()(Embedding(cat_dims[col], int(embedding_size))(input_cat))
        embeddings.append(embedding)

    input_num = Input(shape=(len(numerical_features),), name="numerical_input")
    inputs.append(input_num)

    # Combine all input layers
    x = Concatenate()(embeddings + [input_num])

    # Deep layers
    x = Dense(n_units_1, activation='relu')(x)
    x = Dropout(dropout)(x)
    x = Dense(n_units_2, activation='relu')(x)
    x = Dropout(dropout)(x)

    # Output layer
    output = Dense(1, activation='sigmoid')(x)

    model = Model(inputs=inputs, outputs=output)

    # Compile the model with the specified optimizer and learning rate
    optimizer_instance = getattr(keras.optimizers, optimizer)(learning_rate=learning_rate)
    model.compile(optimizer=optimizer_instance, loss='binary_crossentropy', metrics=['AUC'])

    return model


# --- 1. Custom Transformer to Fix Keras Input Error ---
# ======================================================
# This class is essential to fix the error you encountered.
class KerasInputTransformer(BaseEstimator, TransformerMixin):
    """
    Transforms a standard DataFrame into the dictionary format required by our Keras models,
    making it compatible with scikit-learn's cross_val_score.
    """
    def __init__(self, cat_features, num_features):
        self.cat_features = cat_features
        self.num_features = num_features

    def fit(self, X, y=None):
        return self # No fitting necessary for this transformation

    def transform(self, X):
        # The input X to this transformer inside a scikit-learn pipeline is a NumPy array.
        # We must convert it back to a DataFrame using the original column order.
        X_df = pd.DataFrame(X, columns=self.cat_features + self.num_features)

        keras_inputs = {col: np.array(X_df[col]) for col in self.cat_features}
        keras_inputs['numerical_input'] = X_df[self.num_features].values
        return keras_inputs

def create_wide_deep_model(n_units_1=128, n_units_2=64, dropout=0.3, optimizer='Adam', learning_rate=0.001):
    """
    Factory function to build the Wide & Deep Keras model.
    """
    # Assumes 'categorical_features', 'cat_dims', 'numerical_features', 'Input',
    # 'Flatten', 'Embedding', 'Concatenate', 'Dense', 'Dropout', 'Model', and 'keras' are defined/imported.

    inputs, embeddings = [], []
    for col in categorical_features:
        input_cat = Input(shape=(1,), name=col)
        inputs.append(input_cat)
        embedding_size = min(np.ceil(cat_dims[col]/2), 50)
        embedding = Flatten()(Embedding(cat_dims[col], int(embedding_size))(input_cat))
        embeddings.append(embedding)

    input_num = Input(shape=(len(numerical_features),), name="numerical_input")
    inputs.append(input_num)

    # --- Deep Path ---
    # This path learns complex, abstract patterns through non-linear transformations.
    deep_path = Concatenate()(embeddings + [input_num])
    deep_path = Dense(n_units_1, activation='relu')(deep_path)
    deep_path = Dropout(dropout)(deep_path)
    deep_path = Dense(n_units_2, activation='relu')(deep_path)

    # --- Wide Path ---
    # This path acts as a linear model, learning simple, direct feature interactions.
    wide_path = Concatenate()(embeddings + [input_num])

    # --- Combination ---
    # The outputs of both paths are combined to make the final prediction.
    combined = Concatenate()([wide_path, deep_path])
    output = Dense(1, activation='sigmoid')(combined)

    model = Model(inputs=inputs, outputs=output)

    # Compile the model with the specified optimizer and learning rate
    optimizer_instance = getattr(keras.optimizers, optimizer)(learning_rate=learning_rate)
    model.compile(optimizer=optimizer_instance, loss='binary_crossentropy', metrics=['AUC'])

    return model

# --- 2. Keras Objective Function ---
# ==============================================
def keras_objective(trial, model_builder_func, X, y):
    """Objective function for Keras models, now using a Pipeline to prevent errors."""
    n_units_1 = trial.suggest_int('n_units_1', 64, 512)
    n_units_2 = trial.suggest_int('n_units_2', 32, 256)
    dropout = trial.suggest_float('dropout', 0.1, 0.5)
    optimizer = trial.suggest_categorical('optimizer', ['Adam', 'RMSprop'])
    learning_rate = trial.suggest_float('learning_rate', 1e-4, 1e-2, log=True)
    batch_size = trial.suggest_categorical('batch_size', [64, 128, 256])

    # Create the KerasClassifier with the suggested hyperparameters
    keras_classifier = KerasClassifier(
        model=model_builder_func,
        model__n_units_1=n_units_1, model__n_units_2=n_units_2, model__dropout=dropout,
        model__optimizer=optimizer, model__learning_rate=learning_rate,
        batch_size=batch_size, epochs=50, validation_split=0.2,
        callbacks=[EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)],
        verbose=0
    )

    # <-- FIX: Wrap the Keras model in a pipeline with the input transformer.
    # This pipeline now accepts a standard DataFrame/Numpy array.
    pipeline = Pipeline([
        ('keras_input_transformer', KerasInputTransformer(categorical_features, numerical_features)),
        ('classifier', keras_classifier)
    ])

    return cross_val_score(pipeline, X, y, cv=3, scoring='roc_auc', n_jobs=-1).mean()

# --- 3. Main Execution Loop (Upgraded) ---
# =========================================
#SPECIAL_MODEL_NAMES = ['Keras NN', 'Wide & Deep', 'XGB on AE Features']
SPECIAL_MODEL_NAMES = ['Wide & Deep']
INTENSIVE_OPTUNA_TRIALS = 15
N_TOP_MODELS_TO_TUNE = 4
top_models_names = results_df_sorted['Model'].head(N_TOP_MODELS_TO_TUNE).tolist()
#for model_to_include in SPECIAL_MODEL_NAMES:
#    if model_to_include in results_df_sorted['Model'].values and model_to_include not in top_models_names:
#        top_models_names.append(model_to_include)

print(f"✅ Models identified for intensive tuning: {top_models_names}")

# Create a lookup for initial scores for later comparison.
initial_scores = results_df_sorted.set_index('Model')['Score'].to_dict()
best_tuned_params = {row['Model']: row['Best_Parameters'] for index, row in results_df_sorted.iterrows()}

for model_name in top_models_names:
    print(f"\n--- Running Intensive Optuna for {model_name} ---")
    study = optuna.create_study(direction='maximize')

    try:
        if model_name in ['Wide & Deep']:
            builder_func = create_wide_deep_model
            study.optimize(lambda trial: keras_objective(trial, builder_func, X_train, y_train), n_trials=INTENSIVE_OPTUNA_TRIALS)
        elif model_name in tuning_config:
            study.optimize(lambda trial: universal_sklearn_objective(trial, model_name), n_trials=INTENSIVE_OPTUNA_TRIALS)
        else:
            print(f"⚠️ No tuning configuration found for {model_name}. Skipping.")
            continue

        # --- UPGRADED RESULTS REPORTING ---
        new_score = study.best_value
        new_params = study.best_params
        old_score = initial_scores.get(model_name, 0.0)

        print(f"Initial Score: {old_score:.4f}  |  New Tuned Score: {new_score:.4f}")
        print("Best Parameters Found:")
        for key, value in new_params.items():
            print(f"  - {key}: {value}")

        # <-- NEW: Conditionally update parameters only if the new score is better.
        if new_score > old_score:
            best_tuned_params[model_name] = new_params
            print(f"✅ Improvement found! New parameters for '{model_name}' have been updated.")
        else:
            print(f"⚠️ No improvement. Keeping the original parameters for '{model_name}'.")

    except Exception as e:
        print(f"❌ An error occurred during tuning for {model_name}: {e}")

print("\n✅ Intensive tuning loop completed.")
```

## 11. Final Ensemble Construction with Top 4 Models
Once the best hyperparameters are found out, we are then taking them and building the final model as an ensemble of them all.

```
# --- Custom Transformer for the Autoencoder Pipeline ---

class AutoencoderTransformer(BaseEstimator, TransformerMixin):
    """
    A scikit-learn compatible transformer that fits an autoencoder and uses
    the trained encoder to transform data.
    """
    def __init__(self, encoding_dim=32, ae_optimizer='adam', epochs=50, batch_size=256):
        self.encoding_dim = encoding_dim
        self.ae_optimizer = ae_optimizer
        self.epochs = epochs
        self.batch_size = batch_size

    def fit(self, X, y=None):
        # Build and train the autoencoder on the provided data (X_train_keras)
        input_dim = X.shape[1]
        input_layer = Input(shape=(input_dim,))
        encoder = Dense(128, activation='relu')(input_layer)
        encoder = Dense(self.encoding_dim, activation='relu')(encoder)
        decoder = Dense(128, activation='relu')(encoder)
        decoder = Dense(input_dim, activation='sigmoid')(decoder)

        autoencoder = Model(inputs=input_layer, outputs=decoder)
        autoencoder.compile(optimizer=self.ae_optimizer, loss='mean_squared_error')
        autoencoder.fit(X, X, epochs=self.epochs, batch_size=self.batch_size, validation_split=0.2,
                        callbacks=[EarlyStopping(monitor='val_loss', patience=5)], verbose=0)

        # Store the trained encoder model
        self.encoder_model_ = Model(inputs=input_layer, outputs=encoder)
        return self

    def transform(self, X):
        # Use the fitted encoder to transform new data
        return self.encoder_model_.predict(X, verbose=0)


print("--- Selecting the Top 4 Models for the Final Ensemble ---")
ensemble_candidates = results_df_sorted['Model'].head(4).tolist()
print(f"✅ Selected candidates: {ensemble_candidates}")

# This map remains useful for scikit-learn models
model_map = {
    'XGBoost': XGBClassifier, 'LightGBM': LGBMClassifier, 'CatBoost': CatBoostClassifier,
    'Logistic Regression': LogisticRegression, 'Ridge Classifier': RidgeClassifier,
    'SGD Classifier (Lasso)': lambda **kwargs: SGDClassifier(loss='log_loss', **kwargs),
    'LDA': LinearDiscriminantAnalysis, 'Support Vector Machine': lambda **kwargs: SVC(probability=True, **kwargs),
    'Decision Tree': DecisionTreeClassifier, 'Random Forest': RandomForestClassifier, 'Extra Trees': ExtraTreesClassifier,
    'Gradient Boosting': GradientBoostingClassifier, 'KNN': KNeighborsClassifier, 'Gaussian NB': GaussianNB, 'MLP': MLPClassifier
}

# 2. Rebuild Estimators as Full Imblearn Pipelines
# ==================================================
# This is the key change. We build a complete ImbPipeline for each estimator.

stacking_estimators = []
for model_name in ensemble_candidates:
    estimator_name = model_name.lower().replace(' & ', '_').replace(' ', '_')
    params = best_tuned_params.get(model_name, {})

    if model_name in model_map:
        classifier = model_map[model_name](**params)

        # Standardize common parameters
        if 'random_state' in classifier.get_params(): classifier.set_params(random_state=42)
        if 'n_jobs' in classifier.get_params(): classifier.set_params(n_jobs=-1)
        if 'verbose' in classifier.get_params(): classifier.set_params(verbose=0)

        # Determine the correct preprocessor
        model_type_str = str(type(classifier)).lower()
        if 'nb' in model_type_str: preprocessor = preprocessor_nb
        elif any(tn in model_type_str for tn in ['tree', 'xgb', 'lgbm', 'catboost', 'forest']): preprocessor = preprocessor_tree
        else: preprocessor = preprocessor_linear

        # <-- FIX: Create an ImbPipeline that includes preprocessing AND sampling
        estimator = ImbPipeline([
            ('preprocessor', preprocessor),
            ('sampler', SMOTEENN(random_state=42)),
            ('classifier', classifier)
        ])

        stacking_estimators.append((estimator_name, estimator))
        print(f"✅ Added '{model_name}' to the ensemble pipeline.")
    else:
        # Keras and other special models are skipped as they have different input requirements
        # that are not compatible with this unified preprocessing pipeline.
        print(f"⚠️ Skipping '{model_name}' from the final ensemble due to unique data input requirements.")


# 3. Train the Final Voting Ensemble
# ====================================
from sklearn.ensemble import VotingClassifier

if stacking_estimators:
    # Use 'soft' voting to average predicted probabilities for better performance
    final_ensemble_model = VotingClassifier(estimators=stacking_estimators, voting='soft', n_jobs=-1)

    print("\n--- Training Final Voting Ensemble ---")

    # Now, we fit the ensemble directly on the original X_train.
    # Each internal pipeline will handle its own preprocessing and sampling.
    final_ensemble_model.fit(X_train, y_train)

    print("\n✅ Final Ensemble Model has been trained successfully.")
else:
    print("\n❌ No compatible models found for the final ensemble. Please review the top models.")
```

## 12. Optimizing the Decision Threshold
This code finds the best decision threshold (instead of the default 0.5) for converting predicted probabilities into class labels, in order to maximize the F1-score and ensure a balance between precision and recall.

```
y_proba_final = final_ensemble_model.predict_proba(X_test)[:, 1]
precision, recall, thresholds = precision_recall_curve(y_test, y_proba_final)
f1_scores = (2 * precision * recall) / (precision + recall)
optimal_idx = np.nanargmax(f1_scores)
optimal_threshold = thresholds[optimal_idx]
print(f"Optimal Threshold: {optimal_threshold:.4f}")
```
